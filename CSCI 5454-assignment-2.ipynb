__(A, 5 points)__ A fixed point of an array $A$, if it exists, is an index $i$ such that $A[i] = i$.
Given a _sorted_ array $A$ of _distinct_ integers, return the index of the fixed point if one exists, or otherwise, return `-1` to signal that no fixed point exists. Your algorithm must be as efficient as possible.

```python
def find_fixed_point(a):
    # YOUR SOLUTION HERE
    #binary search
    n = len(a)
  
    return(binary_search(a, 0, n-1))
def binary_search(a, left, right):
    
    if(left>right):
            return -1
    else:
        
        mid = int((left + right)/2)
        if(mid<left or mid>right):
            return -1
        else:
            if(a[mid]==mid):
                return mid
            elif(mid<a[mid]):
                return(binary_search(a, left, mid-1) )
            else:
                return(binary_search(a, mid+1, right))       
                   
    
    # END
```

for $m = 0, \ldots, n-1$. Here $\omega$ is the $n^{th}$ root of unity and $\overline{\omega} = \omega^{-1}$.

## Solution (Expected Length: 15 lines)



### Answer (Expected: 10 lines)
$$LHS = \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  ({\omega})^{-jm} $$
$$LHS = \sum_{j=0}^{n-1}\left(\sum_{k=0}^{n-1} a_k \omega^{jk-jm}\right)$$ 
$$LHS = \sum_{j=0}^{n-1}\left(\sum_{k=0}^{n-1} a_k \omega^{j(k-m)}\right)$$ 
$$LHS = \sum_{j=0}^{n-1}a_k\sum_{k=0}^{n-1} \omega^{j(k-m)}$$ 
$$LHS = \sum_{j=0}^{n-1}a_k\left(\frac{w^{n(k-m)}-1}{w^{(k-m)}-1}\right)$$
But, for $$k\neq m$$ $$\left(\frac{w^{n(k-m)}-1}{w^{(k-m)}-1}\right) = 0$$ 
and for k=m $$ \sum_{j=0}^{n-1} \omega^{j(0)}=n$$
Therefore, at k=m,
 $$LHS = \sum_{k=0}^{n-1}a_mn$$
 $$LHS = a_mn$$
 Hence,
 $$ \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} = n a_{\color{red}{m}} $$
